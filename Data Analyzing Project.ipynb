# Question 5 Results
Are you more likely to get a job if you have a degree?

* Insights- There are significant relationships with having an Associates, Bachelors, and or a Masters degree with part time and fulltime employment.

```python
df["EdLevel"]
```

```python
df['Employment'] = df['Employment'].str.split(';')
```

```python
# Sample data
data = {
    'EdLevel': [
        'Bachelor’s degree',
        'Master’s degree;Bachelor’s degree',
        'Master’s degree',
        'Bachelor’s degree',
        'Associate degree',
    ],
    'Employment': [
        'Employed, full-time',
        'Employed, part-time;Student, part-time',
        'Employed, full-time',
        'Student, full-time',
        'Employed, part-time',
    ],
}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert the 'EdLevel' column to strings
df['EdLevel'] = df['EdLevel'].astype(str)

# Group similar education levels into broader categories (you can customize this)
def aggregate_education_level(education_level):
    if "Master’s degree" in education_level:
        return "Master’s degree"
    elif "Bachelor’s degree" in education_level:
        return "Bachelor’s degree"
    elif "Associate degree" in education_level:
        return "Associate degree"
    elif "Some college/university study without earning a degree" in education_level:
        return "Some college/university study"
    else:
        return education_level

# Apply the aggregation function to the 'EdLevel' column
df['EdLevel'] = df['EdLevel'].apply(aggregate_education_level)

# Create a cross-tabulation without duplicate labels
cross_tab = pd.crosstab(df['EdLevel'], df['Employment'].explode())

# Calculate the percentage of each combination relative to the total number of respondents
cross_tab_percentage = cross_tab.div(cross_tab.sum(axis=1), axis=0) * 100

# Visualize the data as a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(cross_tab_percentage, annot=True, cmap="YlGnBu", fmt=".1f", cbar=False)
plt.title("Relationship Between Education Level and Employment Status")
plt.xlabel("Employment Status")
plt.ylabel("Education Level")
plt.xticks(rotation=45)
plt.show()
```
